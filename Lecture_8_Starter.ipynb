# Functions

```python
# setup
import pandas as pd
```

```python
# score processor

def get_scores(report_card_dictionary, division_name):
    
    # convert the ditionary into dataframe
    report_card_df = pd.DataFrame(report_card_dictionary)
    
    # add a column called division (e.g A,B,C)
    report_card_df["division"] = division_name
    
    # add a column called total score ( 0 to 400)
    report_card_df["total_score"] = report_card_df["math_scores"] + report_card_df["chemistry_scores"] + report_card_df["physics_scores"] + report_card_df["information_technology_scores"]
    
    # add a column called average score ( 0 to 100)
    report_card_df["average_score"] = report_card_df["total_score"]/4
    
    #add a column called status (pass or fail)
    report_card_df.loc[report_card_df["average_score"]>=35, "status"] = "pass"
    report_card_df.loc[report_card_df["average_score"]<35, "status"] = "fail"
    
    # add character grade column ( A,B,C,D)
    report_card_df.loc[report_card_df["average_score"]>=90, "character_grade"] = "A"
    report_card_df.loc[(report_card_df["average_score"]>=70) & (report_card_df["average_score"]<90), "character_grade"] = "B"
    report_card_df.loc[(report_card_df["average_score"]>=50) & (report_card_df["average_score"]<70), "character_grade"] = "C"
    report_card_df.loc[(report_card_df["average_score"]>=35) & (report_card_df["average_score"]<50), "character_grade"] = "D"
    report_card_df.loc[report_card_df["average_score"]<35, "character_grade"] = "F"
    
    #return the dataframe
    return report_card_df
```

# Get Emails Function


```python
def get_emails(emails_dictionary, division_name):
    
    # converting the dictionary into a dataframe
    emails_df = pd.DataFrame(emails_dictionary)
    
    #creating a column called division
    emails_df["division"] = division_name
    
    #converting all string values in the student email column to lower case
    emails_df["student_email"] = emails_df["student_email"].str.lower()
    
    #return the dataframe
    return emails_df
```
