```python
report_card_dictionary = {"student_names": ["hritik", "rahul","paresh", "meet", "dev", "chintan"],
                         "math_grades": [ 98,89,89 ,65,67,34],
                         "biology_grades": [87,87,87,65,79,67],
                         "physics_grades": [ 67,76,76,34,67,90]
                         
                        }
report_card_dictionary
                        
```

```python
#setup
import pandas as pd
```

```python
# making the report card dictionary into a dataframe
report_card_df = pd.DataFrame(report_card_dictionary)
report_card_df
```

```python
# first 3 rows in the report card dataframe
report_card_df.head(3)
```

```python
# 3 random rows in the report card dataframe
report_card_df.sample(3)
```

# insights 
 what is the average math grade? 73.66
 
 what is the lowest biology grade? 65
 
 what is the highest physics grade? 90
 
 how many scores are present in the math column? 6 
 
 
 

```python
# make a total grade column 
report_card_df["total_grade"]= report_card_df["math_grades"] + report_card_df["biology_grades"] + report_card_df["physics_grades"]


#printed the report card dataframe to see the added column
report_card_df
```

```python
report_card_df["average_grades"]= report_card_df["total_grade"]/ 3
report_card_df
```

```python
# saving the report card insights dataframe
report_card_insights_df = report_card_df.describe().transpose()
report_card_insights_df
```

```python
# saving the report card insights as a CSV file 
report_card_insights_df.to_csv("report card insights New.csv")
```

# Report card Alphabet Grades

```python
report_card_df
```

```python
report_card_df.loc[report_card_df["total_grade"]>=40, "status"] = "pass"
report_card_df.loc[report_card_df["total_grade"]<=40, "status"] = "fail"
report_card_df
```

```python
# Making an Alphabet Grade Column
report_card_df.loc[(report_card_df["average_grades"] >=90), "character_grade"] = "A"
report_card_df.loc[(report_card_df["average_grades"] >=70) & (report_card_df["average_grades"]<90), "character_grade"] = "B"
report_card_df.loc[(report_card_df["average_grades"] >=50) & (report_card_df["average_grades"]<70), "character_grade"] = "C"
report_card_df.loc[(report_card_df["average_grades"] >=35) &  (report_card_df["average_grades"]<50), "character_grade"] = "D"
report_card_df.loc[(report_card_df["average_grades"] <=35), "character_grade"] = "F"


```
